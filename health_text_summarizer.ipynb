```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopwords=list(STOP_WORDS)
stopwords
```

```python
nlp=spacy.load('en_core_web_sm')
nlp
```

```python
doc=nlp(health_text)  
```

```python
tokens=[token.text for token in doc]
print(tokens)
```

```python
punctuation=punctuation + '\n'
punctuation
```

```python
word_frequencies={}
for word in doc:
  if word.text.lower() not in stopwords:
    if word.text.lower() not in punctuation:
      if word.text not in word_frequencies.keys():
        word_frequencies[word.text]=1
      else:
        word_frequencies[word.text]+=1
print(word_frequencies)
```

```python
max_frequency=max(word_frequencies.values())
max_frequency
```

```python
for word in word_frequencies.keys():
  word_frequencies[word]=word_frequencies[word]/max_frequency
print(word_frequencies)
```

```python
sentence_tokens=[sen for sen in doc.sents]
print(sentence_tokens)
```

```python
sentence_scores={}
for sen in sentence_tokens:
  for word in sen:
    if word.text.lower() in word_frequencies.keys():
      if sen not in sentence_scores.keys():
        sentence_scores[sen]=word_frequencies[word.text.lower()]
      else:
        sentence_scores[sen]+=word_frequencies[word.text.lower()]
sentence_scores
```

```python
from heapq import nlargest
```

```python
select_len=int(len(sentence_tokens)*0.3)
select_len
```

```python
summary=nlargest(select_len, sentence_scores, key=sentence_scores.get)
summary
```
